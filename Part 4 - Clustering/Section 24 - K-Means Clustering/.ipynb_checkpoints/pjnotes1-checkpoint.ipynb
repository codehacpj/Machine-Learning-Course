## K-means

clusters the data and helps us get he categories of data which are not so obvious sometimes.<br>


##### Random initializaiton trap:


##### To avoid the trap we use K-means ++ algorithm:
To place centroids in particular way.

#### Choosing the right number of clusters:
Within Clusters sum of squares:<b>(we minimize it but not till the point say number of points = number of clusters --> wcss = 0 Hence we see the graph and see the elbow where the improvement is very good not small change.. and at that point we decide is the appropriate number of clusters. This is elbow method.)</b>



### Now we use the Mall_Customers.csv to segment the clients into segments based on annual income and spending score(more they spend more the score)

to use the elbow method to find the optimal number of clusters, we use wcss method for that we loop with different number of clusters between 1 to 10 and plot it to see where the elbow occurs in the graph. <b>sklearn.cluster KMeans has parameters n_clusters to determine how many clusters, to avoid the random clustering trap we use init = 'KMeans++' another parameter is max_iter for number of iterations before the KMeans converge---> default 300, n_init is number of times the kmeans will run with different initial centroids. </b><br><p>
<b>kmeans.inertia_</b> gives wcss for the kmeans model. When plotted it gives the optimal for this data to be 5 clusters. Now, Kmeans for 5 clusters is made and then <b>y_kmeans is made by kmeans.fit_predict(X) inorder to find the clusters that the data points(people belong to) clusters 0,1,2,3,4 is given by y_kmeans for each customers</b><br><p>
    To plot the clusters we use <b>X[y_kmeans==0,0]---> for X coordinate for 1st cluster and on X-axis--> salary y coordinate is given by X[y_kmeans==0,1] for 1st cluster and y-axis value --->customer score<br>
    hence, <br>
    plt.scatter(X[y_kmeans ==0 ,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
</b> <i>gives point's coordinates for cluster 1 salary X-axis, customer score Y-axis with size of point 100 and color red, with label cluster 1</i>
    <br>
    we do it for 5 clusters and then for centroid of every cluster we use yellow color size 300 to distinguish and then kmeans.cluster_centers_[:,0]-->for X-coordinate and<br>
    kmeans.cluster_centers_[:,1]-->for Y-coordinate

```python
#Kmeans  clustering 

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#Importing the mall dataset

dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:,[3,4]].values

#Elbow method to find the optimal number of clusters

from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300 , n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) #inertia_ method is used to get the WCSS for the model
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# From graph we can see that the number of the optimal clusters is 5 
kmeans = KMeans(n_clusters = 5, init = 'k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X) # to predict the custumers in different clusters

#visualising the clusters

plt.scatter(X[y_kmeans ==0 ,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans ==1 ,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans ==2 ,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[y_kmeans ==3 ,0],X[y_kmeans==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans ==4 ,0],X[y_kmeans==4,1],s=100,c='magenta',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of the clients')
plt.xlabel('Salary')
plt.ylabel('Score')
plt.legend()
plt.show()

```
