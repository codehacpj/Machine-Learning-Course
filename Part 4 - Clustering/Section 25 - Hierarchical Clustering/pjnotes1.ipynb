#### Working of dendograms:
The distance between the two points represent the dissimilarity between them and thus: <br>
At first all are singular clusters, now when we see the distance between two clusters, least dissimilar ones are made to get into one cluster in every step... in dendograms we take the points, join them with a bar high enough to represent the distance between the clusters every time we make them go into one cluster. repeated till they all are in one cluster. Creating the Dendograms, bottom up <br>





#### Using the dendograms to make clusters.
We use the distance or dissimilarity threshold to make the clusters. The number of times the threshold line crosses the vertical lines---> gives us the number of clusters.<br>







#### for optimal number of clusters:
we find the largest distance and then we find the thershold line that crosses the largest distance.


<b>We try to find the largest vertical distance without crossing any horizontal lines of other graphs... the horizontal lines are assumed to expand all the way in the graph. Once we find the longest vertical line without crossing any horizontal line... we can take the threshold horizontal line at any place on it.Number of crossings give us the optimal number of clusters</b>
