# SVR

### Support Vector Regression
Using the regression template we use SVR from sklearn.svm to make the regressor, we use gaussian kernel <b>'rbf'</b>. The prediction is wrong as the the predictor is y=constant horizontal line. <br><p>
The reason behind it being, for most common ML algos in sklearn the feature scaling is integrated part in the algorithm but here SVR is a less common class and thus Feature scaling has to be done before using it. 
<br><p>
After feature scaling while we predict, we can't just use<br><b> regressor.predict(6.5)</b><br>
we have to scale it to the sc_X, but we can't just pass 6.5 we will have to pass [[6.5]] a matrix with one element<br>
    <b>regressor.predict(sc_X.transform(np.array([[6.5]])))</b><br>
but again it will give us scaled value y.<br>
we will have to inverse_transform it to scale it to the sc_y<br>
    <b>y_pred = sc_y.inverse_transform(regressor.predict(sc_X.transform(np.array([[6.5]]))))</b> <br><p>
Again for the continuous visualization we make X_grid and then use it to make the curve look smoother.
