# Decision Trees Regression

x1 and x2 is given... we need to predict y.<br>
decision trees determine the optimal splits in data according to the information gains resulted by the splits. once information gain becomes 0 the splitting stops.<br> <p>
    and then when a new point with value(x1,x2) is predicted to be in one category and thus has one value in the 3rd dim. i.e. the value of y. which is generally the average of the y's of other points already in that category.
    

<b>Its a non linear and non continuous model</b>
we have to visualise it in high resolution as in case of low resolution what happens is... the predictions are connected by straight line and the graph looks like it is overfitted to data. And thus we see a curve overfitted to data not like a decision teree which has to have constant values (average of the other values)between two intervals... this can only be seen with the high resolution plotting with <b>X_grid.</b>
