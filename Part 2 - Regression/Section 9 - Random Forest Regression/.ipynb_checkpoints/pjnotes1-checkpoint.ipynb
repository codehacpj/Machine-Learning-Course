# Random Forests

these are the team of decision trees each making the prediction for our random variables, radom forest takes average of these predictions as a whole... form of ensemble learning. <p>
We have more steps than decision trees so lot more splits of data than when for decision trees. more trees more the more steps and average by them converge to some ultimate average.<b>addition of trees makes it possible for the model not to add steps in proportional amount but to place the steps better suited for the problem and thus make the better prediction as the the average converges to the ultimate average for the the given data section.</b>
